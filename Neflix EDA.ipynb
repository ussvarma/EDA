# Find all the instances where;
  Category is "movie" and Type is "Drama"
   or 
    Category is "TV Show" and Type is "Kids TV"

```python
# 1 st condition
df[(df["Category"]=="Movie") & (df["Type"].str.contains("Dramas"))]
```

```python
# 2 nd condition
df[(df["Category"]=="TV Show") & (df["Type"].str.contains("Kids' TV"))]
```

```python
# or condition in between them
df[((df["Category"]=="Movie") & (df["Type"].str.contains("Dramas"))) | ((df["Category"]=="TV Show") & (df["Type"].str.contains("Kids' TV")))]
```
